```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(transactions).transform(transactions).astype('int')
```

```python
te_ary
```

```python
pd.set_option('display.max_columns',None)
```

```python
data=pd.DataFrame(te_ary,columns=te.columns_)
```

```python
data.head()
```

```python
from mlxtend.frequent_patterns import apriori
ap=apriori(data,min_support=0.02,use_colnames=True)
```

```python
ap
```

```python
from mlxtend.frequent_patterns import association_rules
as_rules=association_rules(ap,metric='lift',min_threshold=1)
```

```python
as_rules
```

```python
rules=as_rules[['antecedents','consequents','support','confidence','lift']]
```

```python
rules
```

```python
sns.scatterplot(x='support',y='confidence',data=rules,hue='support',c='red')
plt.title('Association rules')
plt.show()
```

```python
rules['consequents']=rules['consequents'].apply(lambda x: list(x)[0]).astype('unicode')
rules['antecedents']=rules['antecedents'].apply(lambda x: list(x)[0]).astype('unicode')
```
